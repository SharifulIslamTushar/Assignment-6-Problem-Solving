```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#As CustomerID is mandatory for tracing the data. So, those data are deleted from the dataframe which do not have any CustomerID

df = df[df['CustomerID'].notna()]
```

```python
#As there is negative value in Quantity which can do misleading, so we only take the positive value and zero

df=df[(df.Quantity>=0)]
```

```python
df.describe().T
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df.Country.drop_duplicates()) 
df.Country = le.transform(df.Country)
df
```

```python
cols=['Quantity','UnitPrice','Country','CustomerID']
```

```python
df1=df.head(1000)
```

```python
sns.pairplot(df1[cols], kind="reg")
```

```python
df.corr()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
points = pca.fit_transform(df[cols])
```

```python
df['PC_1'] = points[:,0]
df['PC_2'] = points[:,1]
```

```python
df
```

```python
sum(pca.explained_variance_ratio_)*100
```

```python
plt.figure(1 , figsize = (20, 20))
sns.scatterplot(x='PC_1', y='PC_2', data=df, s=50)
```

```python
# Let's find right fit for K
from sklearn.cluster import KMeans
import numpy as np
# Within Cluster Sum of Squared Errors (WCSS) 
wcss = []
max_range = 16
for i in range(1, max_range):
    km = KMeans(n_clusters = i, init = 'k-means++')
    km.fit(df[cols])
    wcss.append(km.inertia_)
    
plt.figure(1 , figsize = (7, 4))
plt.plot(np.arange(1 , max_range) , wcss , 'o')
plt.plot(np.arange(1 , max_range) , wcss , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('WCSS')
plt.show()
```

```python
#The elbow method shows that cluster should be 3

algorithm = KMeans(n_clusters = 3) 
algorithm.fit(df[cols])
grp = algorithm.labels_
df['group'] = grp
df
```

```python
plt.figure(1 , figsize = (8, 6))
sns.scatterplot(x='PC_1', y='PC_2', hue='group', data=df, s=100, palette="Set1")
```

```python
grp0 = df[df["group"]==0].copy()
grp1 = df[df["group"]==1].copy()
grp2 = df[df["group"]==2].copy()
```
